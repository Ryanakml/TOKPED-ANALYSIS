* Nama

* Deskripsi
* Image (Utama)
* Link (full link bukan share link)

Data itu kemudian akan dimasukkan ke dalam file xls/sheets dengan format yang sudah ditentukan.

Total data yang dicari sekitar 500 parfum lokal yang valid.

##  1. IMPORT MODULE

```python
import requests as re
import time
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import pandas as pd
```

## 2. AMBIL PARFUM LOKAL PAGE DAN PARSING

```python
url = 'https://www.tokopedia.com/search?st=&q=parfum&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&navsource='

driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)

konten_html = driver.page_source

driver.quit()

soup = BeautifulSoup(konten_html, 'html.parser')
```

```python
url_awal = 'https://www.tokopedia.com'
link = soup.find_all('a', class_='css-gcvz9')

href_parfum_lokal = '/search?navsource=thematic&q=parfum&source=search&srp_component_id=03.79.01.00&srp_disco_url=beli-lokal&srp_page_id=21239&srp_page_title=Beli+Lokal'

parfum_lokal = url_awal + href_parfum_lokal
parfum_lokal
```

```python
driver = webdriver.Chrome()
driver.get(parfum_lokal)
time.sleep(2)

konten_html = driver.page_source

driver.quit()

soup = BeautifulSoup(konten_html, 'html.parser')
```

### NAMA

```python
i = 1
nama_produk = []
gambar_produk = []
link_produk = []
terjual_produk = []
lokasi_produk = []
toko_produk = []

for five in soup.find_all('div', class_='css-jza1fo'):
    for area in five:
        nama = area.find('div', class_='prd_link-product-name').get_text()
        gambar = area.find('img', class_='css-1q90pod')['src']
        link = area.find('a')['href']
        terjual = area.find('span', class_='prd_label-integrity').get_text()
        lokasi = area.find('span', class_='prd_link-shop-loc').get_text()
        toko = area.find('span', class_='prd_link-shop-name').get_text()
        
        nama_produk.append(nama)
        gambar_produk.append(gambar)
        link_produk.append(link)
        terjual_produk.append(terjual)
        lokasi_produk.append(lokasi)
        toko_produk.append(toko)
```

```python
lokasi
```

```python
dic = {
    'Nama Barang' : nama_produk,
    'Brand' : toko_produk,
    'Lokasi' : lokasi_produk,
    'Terjual' : terjual_produk,
    'Image Url' : gambar_produk,
    'Link Produk' : link_produk,
    'Deskripsi' : deskripsi
    
}
```

```python
df = pd.DataFrame(dic)
df
```

```python
df.to_csv('produk.csv', index=False)
```
